```python
import os
import cv2
from PIL import Image  # pip install pillow
import numpy as np    # pip install numpy
 
def train_classifier(data_dir):
    path = [os.path.join(data_dir, f) for f in os.listdir(data_dir)]
     
    faces = []
    ids = []
     
    for image in path:
        img = Image.open(image).convert('L')  # Convert image to grayscale
        imageNp = np.array(img, 'uint8')
        # Assume the ID is encoded in the filename as described
        id = int(os.path.split(image)[1].split(".")[1])
         
        faces.append(imageNp)
        ids.append(id)
         
    ids = np.array(ids)
     
    # Initialize and train face recognizer
    clf = cv2.face.LBPHFaceRecognizer_create()
    clf.train(faces, ids)
    clf.write("classifier.xml")

# Provide the path to the dataset
train_classifier(r"C:\Users\SHIVANSH SHARMA\Desktop\hackathons\bvp_microsoft\face_recognizer")
```

```python
import cv2
import speech_recognition as sr
import pyttsx3
import threading

def draw_boundary(img, classifier, scaleFactor, minNeighbors, color, text, clf, name_id_mapping):
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    features = classifier.detectMultiScale(gray_img, scaleFactor, minNeighbors)
    names = []
    for (x, y, w, h) in features:
        cv2.rectangle(img, (x, y), (x + w, y + h), color, 2)
        id, pred = clf.predict(gray_img[y:y + h, x:x + w])
        confidence = int(100 * (1 - pred / 300))
        name = name_id_mapping.get(id, "UNKNOWN")
        if confidence > 70:
            cv2.putText(img, name, (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 1, cv2.LINE_AA)
        else:
            cv2.putText(img, "UNKNOWN", (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 1, cv2.LINE_AA)
        names.append(name)
    return img, names

def load_name_id_mapping():
    return {1: "Shivansh", 2: "Raihan", 3: "Reet Dua"}

def speak(text):
    engine = pyttsx3.init()
    engine.say(text)
    engine.runAndWait()

def continuous_listen(recognizer, microphone, commands_queue):
    while True:
        with microphone as source:
            recognizer.adjust_for_ambient_noise(source)
            audio = recognizer.listen(source)
        try:
            command = recognizer.recognize_google(audio).lower()
            commands_queue.append(command)
        except sr.UnknownValueError:
            continue
        except sr.RequestError:
            speak("API unavailable, check your internet connection.")
            break

def video_processing(commands_queue):
    faceCascade = cv2.CascadeClassifier(r"C:\Users\SHIVANSH SHARMA\Downloads\Computer Vision Masterclass-20240427T094259Z-001\Computer Vision Masterclass\Cascades\haarcascade_frontalface_default.xml")
    clf = cv2.face.LBPHFaceRecognizer_create()
    clf.read("classifier.xml")
    name_id_mapping = load_name_id_mapping()
    active = False

    while True:
        if commands_queue:
            command = commands_queue.pop(0)

            if "hello cypher" in command:
                active = True
                speak("Hello, how can I assist you?")

            elif "goodbye cypher" in command:
                active = False
                cv2.destroyAllWindows()
                speak("Goodbye! Say 'Hello Cypher' to reactivate.")

            if active and "who is in front of me" in command:
                video_capture = cv2.VideoCapture(0)
                ret, img = video_capture.read()
                if not ret:
                    speak("Error reading from camera.")
                    continue
                img, names = draw_boundary(img, faceCascade, 1.3, 6, (255, 255, 255), "Face", clf, name_id_mapping)
                cv2.imshow("Face Detection", img)
                if names:
                    response = "The person in front of you is " + ", ".join(set(names))
                    speak(response)
                else:
                    speak("No faces detected.")
                video_capture.release()
                cv2.destroyAllWindows()

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

# Initialize recognizer and microphone
recognizer = sr.Recognizer()
microphone = sr.Microphone()
commands_queue = []

# Start the voice command listening thread
listening_thread = threading.Thread(target=continuous_listen, args=(recognizer, microphone, commands_queue))
listening_thread.start()

# Start the video processing in the main thread
video_processing(commands_queue)
```
